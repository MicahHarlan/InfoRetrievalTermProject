```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_precision = df['Precision']
```

```python
df_precision = df_precision[df_precision != 0]
```

```python
# plt.hist(df_precision, bins=5)
# plt.xlabel('Precision')
# plt.ylabel('Frequency')
# plt.title('Precision Distribution')
# plt.show()
sns.histplot(df_precision, bins=5, kde=True)
plt.xlabel('Histogram of Precision on LLM predictions')
plt.ylabel('Count')
plt.title('Precision' + "\n" + "Average: " + str(round(df_precision.mean(), 2)) + " | " + "Median: " + str(round(df_precision.median(), 2)))
plt.savefig('llm-precision.png')
plt.show()
```

```python
df_clustering = pd.read_csv('cluster_similarity.csv')
```

```python
df_clustering = df_clustering[df_clustering['Precision'] != 0]
```

```python
sns.histplot(df_clustering['Precision'], bins=5, kde=True)
plt.xlabel('Histogram of Precision on Clustering predictions')
plt.ylabel('Count')
plt.title('Precision' + "\n" + "Average: " + str(round(df_clustering['Precision'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering['Precision'].median(), 2)))
plt.savefig('clustering-precision.png')
plt.show()
```

```python
sns.histplot(df_clustering['Recall'], bins=5, kde=True)
plt.xlabel('Histogram of Recall on Clustering predictions')
plt.ylabel('Count')
plt.title('Recall' + "\n" + "Average: " + str(round(df_clustering['Recall'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering['Recall'].median(), 2)))
plt.savefig('clustering-recall.png')
plt.show()
```

```python
df_clustering_1 = pd.read_csv('tfidf_similarity.csv')
```

```python
df_clustering_1 = df_clustering_1[df_clustering_1['Precision'] != 0]
```

```python
sns.histplot(df_clustering_1['Precision'], bins=5, kde=True)
plt.xlabel('Histogram of Precision on Clustering predictions')
plt.ylabel('Count')
plt.title('Precision' + "\n" + "Average: " + str(round(df_clustering_1['Precision'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering_1['Precision'].median(), 2)))
plt.savefig('tfidf-precision.png')
plt.show()
```

```python
sns.histplot(df_clustering_1['Recall'], bins=5, kde=True)
plt.xlabel('Histogram of Recall on Clustering predictions')
plt.ylabel('Count')
plt.title('Recall' + "\n" + "Average: " + str(round(df_clustering_1['Recall'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering_1['Recall'].median(), 2)))
plt.savefig('tfidf-recall.png')
plt.show()
```

```python
# Subplot of two precision histograms
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(df_clustering['Precision'], bins=5, kde=True, ax=axs[0])
axs[0].set_xlabel('K-means')
axs[0].set_ylabel('Count')
axs[0].set_title('Precision' + "\n" + "Average: " + str(round(df_clustering['Precision'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering['Precision'].median(), 2)))
    
sns.histplot(df_clustering_1['Precision'], bins=5, kde=True, ax=axs[1])
axs[1].set_xlabel('TF-IDF')
axs[1].set_ylabel('Count')
axs[1].set_title('Precision' + "\n" + "Average: " + str(round(df_clustering_1['Precision'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering_1['Precision'].median(), 2)))
    
plt.savefig('precision-comparison.png')
plt.show()
```

```python
# Subplot of two recall histograms
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(df_clustering['Recall'], bins=5, kde=True, ax=axs[0])
axs[0].set_xlabel('K-means')
axs[0].set_ylabel('Count')
axs[0].set_title('Recall' + "\n" + "Average: " + str(round(df_clustering['Recall'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering['Recall'].median(), 2)))

sns.histplot(df_clustering_1['Recall'], bins=5, kde=True, ax=axs[1])
axs[1].set_xlabel('TF-IDF')
axs[1].set_ylabel('Count')
axs[1].set_title('Recall' + "\n" + "Average: " + str(round(df_clustering_1['Recall'].mean(), 2)) + " | " + "Median: " + str(round(df_clustering_1['Recall'].median(), 2)))

plt.savefig('recall-comparison.png')
plt.show()
```
