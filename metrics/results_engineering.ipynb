```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.concat([pd.read_csv('../Flask/llm_results.csv'), pd.read_csv('../Flask/llm_results_1.csv')])
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_precision = df['Precision']
```

```python
df_precision = df_precision[df_precision != 0]
```

```python
# plt.hist(df_precision, bins=5)
# plt.xlabel('Precision')
# plt.ylabel('Frequency')
# plt.title('Precision Distribution')
# plt.show()
sns.histplot(df_precision, bins=5, kde=True)
plt.xlabel('Histogram of Precision on LLM predictions')
plt.ylabel('Count')
plt.title('Precision' + "\n" + "Average: " + str(round(df_precision.mean(), 2)) + " | " + "Median: " + str(round(df_precision.median(), 2)))
plt.show()
```

```python
df_recall = df['Recall']
```

```python
df_recall = df_recall[df_recall != 0]
```

```python
sns.histplot(df_recall, bins=5, kde=True)
plt.xlabel('Histogram of Recall on LLM predictions')
plt.ylabel('Count')
plt.title('Recall' + "\n" + "Average: " + str(round(df_recall.mean(), 2)) + " | " + "Median: " + str(round(df_recall.median(), 2)))
plt.show()
```
