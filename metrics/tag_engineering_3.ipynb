```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./ml-25m/genome-scores.csv')
```

```python
df = df[df['relevance'] > 0.5]
df.sample(frac=0.001)
df = df[df['movieId'] > 25000]
```

```python
# Scatter seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
sns.scatterplot(data=df.sample(frac=0.001), x="movieId", y="tagId", hue="relevance")
plt.title('Tags for Movies with relevance > 0.5')
plt.tight_layout()
plt.savefig('tags_for_movies.png')
plt.show()
```
